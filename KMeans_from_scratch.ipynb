```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Funções do K-Means do Zero
import random
import math

def euclidean_distance(point1, point2):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(point1, point2)))

def initialize_centroids(data, k):
    return random.sample(data, k)

def assign_clusters(data, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for point in data:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        cluster_index = distances.index(min(distances))
        clusters[cluster_index].append(point)
    return clusters

def calculate_new_centroids(clusters):
    centroids = []
    for cluster in clusters:
        if cluster:
            centroids.append([sum(dim) / len(dim) for dim in zip(*cluster)])
        else:
            centroids.append([])
    return centroids

def has_converged(old_centroids, new_centroids, tolerance=1e-4):
    for old, new in zip(old_centroids, new_centroids):
        if euclidean_distance(old, new) > tolerance:
            return False
    return True

def kmeans(data, k, max_iterations=100):
    centroids = initialize_centroids(data, k)
    for iteration in range(max_iterations):
        clusters = assign_clusters(data, centroids)
        new_centroids = calculate_new_centroids(clusters)
        if has_converged(centroids, new_centroids):
            print(f"Convergiu após {iteration + 1} iterações")
            break
        centroids = new_centroids
    return clusters, centroids

# 1. Carregar dados do 20 Newsgroups
categories = ['comp.graphics', 'rec.sport.hockey', 'sci.med', 'talk.politics.guns']
newsgroups = fetch_20newsgroups(subset='all', categories=categories, remove=('headers', 'footers', 'quotes'))
texts = newsgroups.data

# 2. Pré-processamento: TF-IDF
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(texts).toarray()

# 3. Reduzir para 2 dimensões (PCA)
pca = PCA(n_components=2, random_state=42)
X_2d = pca.fit_transform(X)

# 4. Aplicar K-Means personalizado
k = 4
clusters, centroids = kmeans(X_2d.tolist(), k)

# 5. Visualizar os clusters
plt.figure(figsize=(10, 7))

# Plotar os pontos em cada cluster
for cluster_index, cluster in enumerate(clusters):
    cluster_points = list(zip(*cluster))
    plt.scatter(cluster_points[0], cluster_points[1], label=f"Cluster {cluster_index}")

# Plotar os centróides
centroid_points = list(zip(*centroids))
plt.scatter(centroid_points[0], centroid_points[1], c='black', marker='x', s=100, label='Centroids')

plt.title("Clusters gerados pelo K-Means do Zero")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend()
plt.show()
```
